```python
import folium
from folium.plugins import MarkerCluster

# Crear el mapa centrado en las coordenadas promedio
site_map = folium.Map(location=[spacex_df['Lat'].mean(), spacex_df['Long'].mean()], zoom_start=5)

# Crear un objeto MarkerCluster
marker_cluster = MarkerCluster().add_to(site_map)

# Iterar sobre cada fila del DataFrame para añadir marcadores al cluster
for index, record in spacex_df.iterrows():
    # Extraer datos disponibles
    launch_site = record['Launch Site']
    marker_color = record['marker_color']
    launch_class = 'Success' if record['class'] == 1 else 'Failed'

    # Crear un marcador personalizado
    marker = folium.Marker(
        location=[record['Lat'], record['Long']],
        popup=(
            f"Launch Site: {launch_site}<br>"
            f"Class: {launch_class}"
        ),
        icon=folium.Icon(color=marker_color, icon="info-sign")
    )
    # Añadir el marcador al cluster
    marker_cluster.add_child(marker)

# Mostrar el mapa
site_map
```

```python
# TASK 3: Calculate the distances between a launch site to its proximities
# Agrupar datos por sitio de lanzamiento
unique_sites_df = spacex_df.groupby(['Launch Site', 'Lat', 'Long'], as_index=False).first()

# Punto de referencia (ajústalo según tu caso)
reference_lat = 28.562302  # Latitud de referencia
reference_lon = -80.577356  # Longitud de referencia

# Calcular la distancia para sitios únicos
def calculate_distance(lat1, lon1, lat2, lon2):
    R = 6371  # Radio de la Tierra en kilómetros
    lat1, lon1, lat2, lon2 = map(math.radians, [lat1, lon1, lat2, lon2])
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    a = math.sin(dlat / 2)**2 + math.cos(lat1) * math.cos(lat2) * math.sin(dlon / 2)**2
    c = 2 * math.asin(math.sqrt(a))
    return R * c

unique_sites_df['distance_km'] = unique_sites_df.apply(
    lambda row: calculate_distance(reference_lat, reference_lon, row['Lat'], row['Long']), axis=1
)

# Mostrar resultados únicos
unique_sites_df[['Launch Site', 'Lat', 'Long', 'distance_km']]

```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```
